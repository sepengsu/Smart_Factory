### EDA


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV 파일 읽기 (파일명은 'data.csv'로 가정)
df = pd.read_csv('c_1_001.csv', header=None)

# 컬럼명 지정
df.columns = [
    'Force_X (N)', 'Force_Y (N)', 'Force_Z (N)',
    'Vibration_X (g)', 'Vibration_Y (g)', 'Vibration_Z (g)',
    'AE-RMS (V)'
]
df = df[['Vibration_X (g)', 'Vibration_Y (g)', 'Vibration_Z (g)',
    'AE-RMS (V)']]

# 기본 통계 정보 출력
df.describe()
```

```python
# 각 컬럼별 시계열 그래프 그리기
df.plot(subplots=True, layout=(4, 2), figsize=(15, 10), title="Time Series Plot for Each Column")
plt.tight_layout()
plt.show()
```

```python
# 각 컬럼별 히스토그램 그리기
df.hist(figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
# 각 컬럼별 박스플롯 그리기
plt.figure(figsize=(15, 10))
df.boxplot()
plt.tight_layout()
plt.show()
```

```python
# 상관 행렬 계산 및 히트맵 그리기
correlation_matrix = df.corr()
plt.figure(figsize=(10, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
# Pairplot 그리기
sns.pairplot(df,hue='AE-RMS (V)')
plt.suptitle("Pairplot of Variables", y=1.02)
plt.show()
```
